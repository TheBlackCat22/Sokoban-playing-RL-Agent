```python
import Agent
import Warehouse_Env
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
env=Warehouse_Env.Warehouse()

writer = animation.PillowWriter(fps=2)
```

## On-Policy TD Control (SARSA)

```python
env.reset()
```

```python
policy=Agent.sarsa(env)
```

```python
anim=Agent.play_policy(env,policy)
```

```python
HTML(anim.to_jshtml())
```

```python
anim.save('./results/SARSA.gif', writer=writer)
```

## On-Policy Monte Carlo

```python
env.reset()
```

```python
policy=Agent.on_policy_mc(env,episodes=20,verbose=1)
```

```python
anim=Agent.play_policy(env,policy)
```

```python
HTML(anim.to_jshtml())
```

```python
anim.save('./results/onp_mc.gif', writer=writer)
```

In this situation, both Q-learning and Sarsa perform very well with Q-learning performing Slightly better.

Both the monte-carlo methods take a lot of episodes to train.

On-policy monte-carlo returns a stochastic policy instead of the deterministic policies the other learning algorithms return.

Performance Ordering: Q-Learning > SARSA > Off-policy Monte carlo > On-policy Monte carlo
